```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_scores.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Drop Unnamed column

```python
df =df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

# Gender Distribution

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(data = df , x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

# From the above chart we have analised that:
# no. of females are more than males

```python
gb =df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb, annot= True)
plt.title("Relationship between Parents Education and Students Score's")
plt.show()
```

# From the above chart we can conclude:
# the parent's education has great impact on students studies

```python
gb1 =df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1, annot= True)
plt.title("Relationship between ParentsMarital Status and Students Score's")
plt.show()
```

# From the above chart we can conclude:
# the ParentMaritalStatus has no/negligible impact on students scores

# Distribution of ethnic groups 

```python
groupA=df.loc[(df['EthnicGroup']=="groupA")].count()
print(groupA)
```

```python
groupA=df.loc[(df['EthnicGroup']=="group A")].count()
groupB=df.loc[(df['EthnicGroup']=="group B")].count()
groupC=df.loc[(df['EthnicGroup']=="group C")].count()
groupD=df.loc[(df['EthnicGroup']=="group D")].count()
groupE=df.loc[(df['EthnicGroup']=="group E")].count()

l=["group A","group B","group C","group D","group E"]
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
print(mlist)
plt.pie(mlist ,labels=l, autopct="%1.2f%%")
plt.title("Distribution of Ethinic Groups")
plt.show()
```
